```python
import pickle

with open('tweets.pkl', 'rb') as f:
    tweets_dict = pickle.load(f)
```

```python
import matplotlib.pyplot as plt
import matplotlib.dates as mdates

# Convert strings to datetime objects
dates = [tweets_dict[k].date for k in tweets_dict]

# Create a figure and a subplot
fig, ax = plt.subplots()

# Plot a histogram of the dates
ax.hist(dates, bins=15)

# Set the x-axis to display dates
ax.xaxis.set_major_locator(mdates.DayLocator(interval=3))
ax.xaxis.set_major_formatter(mdates.DateFormatter('%Y-%m-%d'))

# Set the y-axis label
ax.set_ylabel('Frequency')

# Show the plot
plt.xlabel('Dates')
plt.ylabel('No. of tweets in dataset')
plt.show()
```

```python
# Ploting distribution of like count
import matplotlib.pyplot as plt

like_counts = [tweets_dict[k].likeCount for k in tweets_dict if tweets_dict[k]]

plt.hist(like_counts, bins=20)
plt.title('Distribution of Like Counts')
plt.xlabel('Number of Likes')
plt.ylabel('Frequency')
plt.yscale('log')
plt.show()
```

```python
# Ploting distribution of view count
import matplotlib.pyplot as plt
import matplotlib as mpl

# set figure size and resolution
mpl.rcParams['figure.figsize'] = (7, 3)
mpl.rcParams['figure.dpi'] = 150

view_counts = [tweets_dict[k].viewCount for k in tweets_dict if tweets_dict[k].viewCount is not None]

plt.hist(view_counts, bins=20)
plt.title('Distribution of View Counts')
plt.xlabel('Number of views')
plt.ylabel('Frequency')
plt.ticklabel_format(style='plain', axis='x')
plt.yscale('log')
plt.show()
```

```python
import matplotlib.pyplot as plt

# Example list of tuples
data = [(tweets_dict[k].viewCount, tweets_dict[k].likeCount) for k in tweets_dict if tweets_dict[k].viewCount is not None]

# Extracting likeCounts and viewCounts as separate lists
viewCounts = [t[0] for t in data]
likeCounts = [t[1] for t in data]

# Creating a scatter plot with reversed x-axis
plt.scatter(viewCounts, likeCounts)

# Adding labels and title
plt.xlabel('View Count')
plt.ylabel('Like Count')
plt.ticklabel_format(style='plain', axis='x')
plt.title('View Count vs. Like Count')

# Displaying the plot
plt.show()
```

```python
max_ratio = 0
max_id = ''
min_ratio = 1
min_id = ''


for k in tweets_dict:
    if tweets_dict[k].viewCount is None:
        continue
    tweet = tweets_dict[k]
    ratio = (tweet.likeCount + 1) / tweet.viewCount
    if ratio > max_ratio:
        max_ratio = ratio
        max_id = k
    if ratio < min_ratio:
        min_ratio = ratio
        min_id = k

print('Most popular tweet:', tweets_dict[max_id].rawContent)
print('Least popular tweet:', tweets_dict[min_id].rawContent)
```
