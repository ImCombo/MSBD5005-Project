```python
import pickle

with open('D:\Downloads\\tweet_combine_221130-230315.pkl', 'rb') as f:
    tweets_dict = pickle.load(f)
```

```python
# Plot the amount of tweet collected overtime but may not be representative of the actual number of tweet
import matplotlib.pyplot as plt
import matplotlib.dates as mdates

# Convert strings to datetime objects
dates = [tweets_dict[k].date for k in tweets_dict]

# Create a figure and a subplot
fig, ax = plt.subplots()

# Plot a histogram of the dates
ax.hist(dates, bins=50)

# Set the x-axis to display dates
ax.xaxis.set_major_locator(mdates.DayLocator(interval=15))
ax.xaxis.set_major_formatter(mdates.DateFormatter('%Y-%m-%d'))

# Set the y-axis label
ax.set_ylabel('Frequency')

# Show the plot
plt.xlabel('Dates')
plt.ylabel('Tweets in dataset')
plt.show()
```

```python
# Ploting distribution of like count
import matplotlib.pyplot as plt

like_counts = [tweets_dict[k].likeCount for k in tweets_dict if tweets_dict[k]]

plt.hist(like_counts, bins=50)
plt.title('Distribution of Like Counts')
plt.xlabel('Number of Likes')
plt.ylabel('Frequency')
plt.yscale('log')
plt.show()
```

```python
# Ploting distribution of view count
import matplotlib.pyplot as plt

view_counts = [tweets_dict[k].viewCount for k in tweets_dict if tweets_dict[k].viewCount is not None]

plt.hist(view_counts, bins=50)
plt.title('Distribution of View Counts')
plt.xlabel('Number of views')
plt.ylabel('Frequency')
plt.ticklabel_format(style='plain', axis='x')
plt.yscale('log')
plt.show()
```

```python
import matplotlib.pyplot as plt

# Example list of tuples
data = [(tweets_dict[k].viewCount, tweets_dict[k].likeCount) for k in tweets_dict if tweets_dict[k].viewCount is not None]

# Extracting likeCounts and viewCounts as separate lists
viewCounts = [t[0] for t in data]
likeCounts = [t[1] for t in data]

# Creating a scatter plot with reversed x-axis
plt.scatter(viewCounts, likeCounts)

# Adding labels and title
plt.xlabel('View Count')
plt.ylabel('Like Count')
plt.ticklabel_format(style='plain', axis='x')
plt.title('View Count vs. Like Count')

# Displaying the plot
plt.show()
```

```python
import matplotlib.pyplot as plt

# Example list of tuples
like_ratio = [tweets_dict[k].likeCount / tweets_dict[k].viewCount for k in tweets_dict if tweets_dict[k].viewCount is not None]

plt.hist(like_ratio, bins=50)
plt.xlabel('Like to View Ratio')
plt.ylabel('Frequency')
plt.ticklabel_format(style='plain', axis='x')
plt.yscale('log')
plt.show()
```

```python
max_ratio = 0
max_id = ''
min_ratio = 1
min_id = ''


for k in tweets_dict:
    tweet = tweets_dict[k]
    view_count = tweet.viewCount
    like_count = tweet.likeCount
    if view_count is None or view_count < 10000:
        continue
    ratio = (like_count + 1) / (view_count + 1)
    if ratio > max_ratio:
        max_ratio = ratio
        max_id = k
    if ratio < min_ratio:
        min_ratio = ratio
        min_id = k

print('Most popular tweet:', tweets_dict[max_id].rawContent, tweets_dict[max_id].url)
print('Least popular tweet:', tweets_dict[min_id].rawContent, tweets_dict[min_id].url)
```
